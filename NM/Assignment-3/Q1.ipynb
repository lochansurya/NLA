```python
"""definition of the newton's method"""
import numpy as np
import matplotlib.pyplot as plt
import math

def Newton(f, df, z0, tol):
    """Newton's method"""
    z = z0-f(z0)/df(z0)
    while (abs(z-z0)/abs(z0)) > tol:
        z = z - f(z)/df(z)
    return z
```

```python
n=4000
z=np.linspace(-2.5,1,n)
Y=np.linspace(-1.5,1.5,n)
```

```python
from numba import njit
@njit
def Mandelbrot(c, maxiter):
    """Mandalbrot set"""
    z = 0
    iter=1
    while(abs(z) < 2 and iter <= maxiter):
        z = z**2 + c
        iter+=1
        
    return iter
```

```python
Mandelbrotset=np.array([Mandelbrot(z+1j*y,100)for z in z for y in Y]).reshape(n,n)
print(Mandelbrotset)
```

```python
"""save the mandelbrot set"""
plt.imshow(Mandelbrotset.T, extent=[-2.5,1,-1.5,1.5])
"""with different colors"""
plt.imshow(Mandelbrotset.T, extent=[-2.5,1,-1.5,1.5], cmap='hot')
plt.savefig('mandelbrot.png')
```

```python
N=3000
X_prime=np.linspace(-2.0,2.0,N)
Y_prime=np.linspace(-2.0,2.0,N)
```

```python
"""define the function f(z)=z^3+1"""
def f(z):
    return z**3+1

"""definition of the first derivative of f(z)"""
def df(z):
    return 3*z**2
```

```python
"""roots of the function f(z)"""
root1=-1
root2=0.5+(math.sqrt(3)/2)*1j
root3=0.5-(math.sqrt(3)/2)*1j
```

```python
"""definition of newton method with arguements: f, df, z0, tol"""

def Newton(f, df, x, y, tol):
    z0=x+1j*y
    z=z0-f(z0)/df(z0)
    iter=0
    MaxIter=100
    while iter<MaxIter:
        iter+=1
        if(abs(z-root1)<tol):
            return 0
        if(abs(z-root2)<tol):
            return 1
        if(abs(z-root3)<tol):
            return 2
        z = z - f(z)/df(z)
    return MaxIter
```

```python
"""call Newton function by the complex numbers generated by X_prime and Y_prime"""
tol=1e-6
Z=np.array([Newton(f, df, x, y, tol) for x in X_prime for y in Y_prime]).reshape(N,N)
```

```python
"""plot the image"""
plt.imshow(Z.T, extent=[-2.0,2.0,-2.0,2.0],cmap='hot')
plt.savefig('newton.png')
```
