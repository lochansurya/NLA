```python
"""Singular Value Decomposition (SVD)"""
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np

# Load image
img = imread('/Users/lochan_n/Desktop/NLA/Assignments/STScI-01H6C42211PD8XXD6A7VQ3330N 1.png')
```

```python
red=img[:,:,0]
green=img[:,:,1]
blue=img[:,:,2]
```

```python
# SVD on each channel
U_r, S_r, V_r = np.linalg.svd(red)
U_g, S_g, V_g = np.linalg.svd(green)
U_b, S_b, V_b = np.linalg.svd(blue)
```

```python
sr=np.diag(S_r)
sg=np.diag(S_g)
sb=np.diag(S_b)
```

```python
print("number of singular values for red channel:",sr.shape[0])
print("number of singular values for green channel:",sg.shape[0])
print("number of singular values for blue channel:",sb.shape[0])
```

```python
j, j= sr.shape  
print("shape of sigma matrix for red channel:",j,j)
```

```python
r=5
v_rank=[]
sigma=[]
Two_norm=[]
Frobenius_norm=[]
euclid_error_list_red=[]
frobenius_error_list_red=[]
euclid_error_list_green=[]
frobenius_error_list_green=[]
euclid_error_list_blue=[]
frobenius_error_list_blue=[]
"""storing the singular values of the three channels in three lists"""
red_singular_values=[]
green_singular_values=[]
blue_singular_values=[]
"""root square error of the singular values from sigma_(r+1) to sigma_(n)"""
rse_residual_error_red=[]
rse_residual_error_green=[]
rse_residual_error_blue=[]
# euclid_error=[]
# frobeinus_error=[]  


while r<j:
    red_modified=U_r[:,:r]@sr[:r,:r]@V_r[:r,:]
    green_modified=U_g[:,:r]@sg[:r,:r]@V_g[:r,:]
    blue_modified=U_b[:,:r]@sb[:r,:r]@V_b[:r,:]
    euclid_error_list_red.append(np.linalg.norm(red-red_modified,ord=2))                    #2-norm red channel
    frobenius_error_list_red.append(np.linalg.norm(red-red_modified,ord='fro'))             #Frobenius norm red channel
    euclid_error_list_green.append(np.linalg.norm(green-green_modified,ord=2))              #2-norm green channel
    frobenius_error_list_green.append(np.linalg.norm(green-green_modified,ord='fro'))       #Frobenius norm green channel
    euclid_error_list_blue.append(np.linalg.norm(blue-blue_modified,ord=2))                 #2-norm blue channel
    frobenius_error_list_blue.append(np.linalg.norm(blue-blue_modified,ord='fro'))          #Frobenius norm blue channel
    X_mod=np.stack((red_modified,green_modified,blue_modified),axis=-1)                     #stacking the modified channels
    red_singular_values.append(S_r[r])
    green_singular_values.append(S_g[r])
    blue_singular_values.append(S_b[r])
    rse_residual_error_red.append(np.sqrt(S_r[r:].T@S_r[r:]))                              #root square error of the singular values from sigma_(r+1) to sigma_(n
    rse_residual_error_green.append(np.sqrt(S_g[r:].T@S_g[r:]))
    rse_residual_error_blue.append(np.sqrt(S_b[r:].T@S_b[r:]))
    plt.figure()                                                                            #plotting the modified image
    plt.imshow(X_mod)                                                                       
    plt.title("r="+str(r))
    plt.savefig("r="+str(r)+".png")
    plt.show()
    #number of entries in the matrix
    entries=((shape_x*r)+r+(r*shape_y))*shape_z
    print("No. of entries transmitted for compressed image of rank",r,"=",entries)
    print("The compressed image size is ",entries/total_elements_of_the_image_*100,"%"," of the original image size")
    v_rank.append(r)
    sigma.append(r*shape_x+shape_y*r+2*r)
    Two_norm.append(np.linalg.norm(red-red_modified,ord=2))
    Frobenius_norm.append(np.linalg.norm(red-red_modified,ord='fro'))
    r=r*2

```
